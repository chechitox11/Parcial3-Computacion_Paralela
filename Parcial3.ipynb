```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Grafica.csv')
df
```

##Se hace la grafica de los tiempos en segundos tomando la info. de .csv que se creo

```python
df.plot()
plt.legend()
plt.show()
```

```python
print("el promedio de tiempo de ejecucion en cython es de:",df["Cython"].mean())
```

```python
print("el promedio de tiempo de ejecucion en python es de:",df["Python"].mean())
```

```python
a=(df["Python"].mean()-df["Cython"].mean())
```

```python
print("La diferencia del promedio del tiempo de ejecucion entre ambos es de:",a)
```

##Se pude apreciar por el cambio tan grande que tuvieron los tiempos de ejecucion, que el cython (siendo una forma de codigo cuya mision es dar un rendimiento mucho mejor, como el encontrado en el lenguaje C) hace una ejecucion mucho más rapida frente al codigo hecho en python3, cabe incluir que la diferencia entre el promedio de los dos baja el tiempo de 5s a 1s
